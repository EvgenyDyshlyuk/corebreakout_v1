```python
import numpy as np
from skimage import io

from corebreakout import CoreColumn
```

```python
img1 = io.imread('../data/column1.jpeg') 
img2 = io.imread('../data/column2.jpeg')

col1 = CoreColumn(img1, top=1.0, base=2.0)
col2 = CoreColumn(img2, top=2.0, base=3.0)

print(col1, col2)
```

# Single column plot

```python
col1.plot(figsize=(7, 25))
```

# Adding adjacent columns

```python
col12 = col1 + col2
col12.plot(figsize=(12, 40))
```

# With `add_mode='fill'`

```python
col1 = CoreColumn(img1, top=1.0, base=2.0, add_tol=1.0)
col2 = CoreColumn(img2, top=3.0, base=4.0)

col12 = col1 + col2
col12.plot(figsize=(12, 60))
```

# With `add_mode='collapse'`

```python
col1 = CoreColumn(img1, top=1.0, base=2.0, add_tol=1.0, add_mode='collapse')
col2 = CoreColumn(img2, top=3.0, base=4.0)

col12 = col1 + col2
col12.plot(figsize=(12, 40))
```

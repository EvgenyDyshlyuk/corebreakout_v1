```python
from pathlib import Path

from corebreakout import defaults, CoreSegmenter
```

```python
# Using images from mini testing dataset
img_path_1 = '../data/two_image_dataset/S00101409.jpeg'
img_path_2 = '../data/two_image_dataset/S00111582.jpeg'

# Made up depth ranges s.t. stacking is allowed
depth_ranges = [[1.0, 3.0], [3.0, 5.0]]
```

```python
# Using default locations. Modify if desired.
model_dir = defaults.MODEL_DIR
weights_path = defaults.CB_MODEL_PATH

# Instantiate with default config/classes/layout kwargs
segmenter = CoreSegmenter(model_dir, weights_path)
```

```python
# Single image --> CoreColumn
single_img_col = segmenter.segment(img_path_1, depth_ranges[0], show=True)

print(single_img_col)
```

```python
# Multiple images --> CoreColumn
combined_imgs_col = segmenter.segment_all([img_path_1, img_path_2], depth_ranges, show=True)

print(combined_imgs_col)
```
